***9 . average weekly response time for each agent*** 

```python
def convert(seconds):
    seconds = seconds % (24 * 3600)
    hour = seconds // 3600
    seconds = seconds%3600
    minutes = seconds // 60
    seconds = seconds%60
    return "%d:%02d:%02d" % (hour, minutes, seconds)

# converting time resolution in seconds
df['Time_responce'] = pd.to_timedelta(df['Average_Response_Time']).astype('timedelta64[s]').astype(int)
df_time= pd.DataFrame(df[df['Average_Rating'] >  0] .groupby(['Agent_Name', 'Week'])['Time_responce'].mean())

# Converting sec back into time 
df_time['Time_response'] = df_time['Time_responce'].apply(convert)
df_time
```

***10 . average weekely resolution time for each agents***

```python
def convert(seconds):
    seconds = seconds % (24 * 3600)
    hour = seconds // 3600
    seconds = seconds%3600
    minutes = seconds // 60
    seconds = seconds%60
    return "%d:%02d:%02d" % (hour, minutes, seconds)

# converting time resolution in seconds
df['Time_resolution'] = pd.to_timedelta(df['Average_Resolution_Time']).astype('timedelta64[s]').astype(int)
df_time= pd.DataFrame(df[df['Average_Rating'] >  0] .groupby(['Agent_Name', 'Week'])['Time_resolution'].mean())

# Converting sec back into time 
df_time['Time_resolution'] = df_time['Time_resolution'].apply(convert)
df_time
```
